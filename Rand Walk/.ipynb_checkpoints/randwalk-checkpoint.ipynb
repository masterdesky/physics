## Rand walk

```python
n_steps = 100000
X = np.zeros(n_steps)
Y = np.zeros(n_steps)
Z = np.zeros(n_steps)
R = np.zeros(n_steps)
```

```python
for i in range(1,n_steps):
    X[i] = np.random.normal() + X[i-1]
    Y[i] = np.random.normal() + Y[i-1]
    Z[i] = np.random.normal() + Z[i-1]
    R[i] = np.sqrt(X[i]**2 + Y[i]**2 + Z[i]**2)

low = np.min([np.min(X), np.min(Y), np.min(Z)])
high = np.max([np.max(X), np.max(Y), np.max(Z)])
```

```python
save = False

nrows=1
ncols=1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*12,nrows*12),
                         subplot_kw={'projection': '3d'})

axislabelsize = 23
axisticksize = 20
labelpad = 25
tickpad = 15
scattersize = 2

axes.set_xlabel('Distance (X)', fontsize=axislabelsize, labelpad=labelpad)
axes.set_ylabel('Distance (Y)', fontsize=axislabelsize, labelpad=labelpad)
axes.set_zlabel('Distance (Z)', fontsize=axislabelsize, labelpad=labelpad)

#axes.set_xlim(low,high)
#axes.set_ylim(low,high)
#axes.set_zlim(low,high)

axes.tick_params(axis='both', which='major',
                 pad=tickpad, labelsize=axisticksize)

axes.plot(X, Y, Z)

plt.show()
```

```python
save = False

nrows=1
ncols=1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*12,nrows*12))

axislabelsize = 23
axisticksize = 20
labelpad = 25
tickpad = 15
scattersize = 2

axes.set_xlabel('Square root of number of steps [$\sqrt{N}$]', fontsize=axislabelsize, labelpad=labelpad)
axes.set_ylabel('Distance from start', fontsize=axislabelsize, labelpad=labelpad)

axes.tick_params(axis='both', which='major',
                 pad=tickpad, labelsize=axisticksize)

axes.plot(np.sqrt(np.linspace(0,n_steps,n_steps)), R)

plt.show()
```
