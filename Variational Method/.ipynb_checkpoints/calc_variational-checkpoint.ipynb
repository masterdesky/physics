Szeretnénk meghatározni egy hidrogénszerű atom energiaszintjeit variációszámítás segítségével. Belátható, hogy az energiaszintek helyét az

$$
E_{0}
=
\frac{\left< \Psi_{0}^{\ast} | \hat{H} | \Psi_{0} \right>}{\left< \Psi_{0}^{\ast} | \Psi_{0} \right>}
=
\frac{\int_{-\infty}^{\infty} \text{d}r\, \Psi_{0}^{\ast} \hat{H} \Psi_{0}}
{\int_{-\infty}^{\infty} \text{d}r\, \Psi_{0}^{\ast} \Psi_{0}}
$$

egyenlet megoldásai adják eredményül. Itt $\Psi_{0}$ a valódi hullámfügvény. Keressük a közelítő hullámfüggvényünket a feladat kiírása szerint az

$$
\phi
\propto
e^{- \alpha r}
$$

alakban. Ahol $\alpha$ egy valós paraméter. Ekkor az energiaszintek értékét a következő egyenlettel közelíthetjük:

$$
E_{\phi}
=
\frac{\left< \phi^{\ast} | \hat{H} | \phi \right>}{\left< \phi^{\ast} | \phi \right>}
=
\frac{\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{H} \phi}
{\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \phi}
$$

Ahol a nevezőben szereplő tag egy normalizációs faktor, míg a számlálóban található tag az alábbi módon fejthető ki:

$$
\left< \phi^{\ast} | \hat{H} | \phi \right>
=
\left< \phi^{\ast} | \hat{T} | \phi \right>
+
\left< \phi^{\ast} | \hat{V} | \phi \right>
=
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{T} \phi
+
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{V} \phi
$$

Ahol a hidrogénszerű atom esetében a kinetikus energia operátora

$$
\hat{T}
=
\frac{\hbar^{2}}{2m} \Delta
$$

a potenciális energia operátora pedig

$$
\hat{V}
=
\frac{1}{2} k r^{2}
$$

Érdekes lehet ábrázolni a potenciál és a próbafüggvény menetét a jobb átláthatóság reményében.

```python
def trial(r, a):
    
    return np.exp(- a * r)
```

```python
def V(r, k):

    return 1/2 * k * r**2
```

```python
r_values = np.arange(-10, 10, 0.01)
k_values = [0.1, 0.2, 0.5, 1, 1.5, 3]
```

```python
nrows = 1
ncols = 2
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*10, nrows*8))

axislabelsize = 25
axisticksize = 15
axislegendsize = 16

colors = [
    'tab:red',
    'tab:green',
    'tab:orange',
    'tab:blue',
    'tab:purple',
    'tab:cyan'
]

for idx, k in enumerate(k_values):
    axes[0].plot(r_values, V(r_values, k), label='Potential with $k = {}$'.format(k),
                 c=colors[idx], lw=3)

k = k_values[-1]
axes[1].plot(r_values, V(r_values, k=k), label='Potential with $k = {}$'.format(k),
                 c='tab:red', lw=3)
axes[1].plot(r_values, trial(r_values, a=0.5), label='Trial function'.format(k),
                 c='tab:green', lw=3)

for i in range(ncols):
    axes[i].set_xlabel('$r$', fontsize=axislabelsize)
    axes[0].set_ylabel('$\\frac{1}{2} k r^{2}$', fontsize=axislabelsize)
    axes[1].set_ylabel('Function values', fontsize=axislabelsize)
    axes[i].tick_params(axis='both', which='major', labelsize=axisticksize)

    axes[i].legend(loc='upper right', fontsize=axislegendsize)

plt.show()
```

Most számítsuk ki ezen $E_{\phi}$ értékét. A normalizációs faktor értéke a própafüggvényünk alapján

$$
\left< \phi^{\ast} | \phi \right>
=
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \cdot \phi
=
\int_{-\infty}^{\infty} \text{d}r\, \left( e^{- \alpha r} \right)^{\ast} \cdot \left( e^{- \alpha r} \right)
$$

Mivel az $e^{- \alpha r}$ kifejezésben szereplő $r$ távolság egy valós mennyiség, ezért annak komplex konjugáltja önmaga lesz. Így felírható az alábbi kifejezés:

$$
\left< \phi^{\ast} | \phi \right>
=
\dots
=
\int_{-\infty}^{\infty} \text{d}r\, \left( e^{- \alpha r} \right) \cdot \left( e^{- \alpha r} \right)
=
\int_{-\infty}^{\infty} \text{d}r\, e^{-2 \alpha r}
\equiv
\int_{0}^{\infty} \text{d}r\, e^{-2 \alpha r}
$$
Mivel $r$ itt - gondolom - egy a hidrogénszerű atom esetében egy gömbi radiális koordináta akar lenni, így elég az integrálokat $0$ és $+\infty$ között elvégezzük, ugyanis $r < 0$ értékek nem értelmezhetőek egy ilyen kép esetén.

Ezt numerikusan megoldhatjuk pl. a `sympy` modul segítségével.

```python
r = sp.Symbol('r')
alpha = sp.Symbol('\\alpha')
```

```python
result = sp.integrate(sp.exp(-2 * alpha * r), (r, 0, sp.oo))
result
```

Azt az eredményt kaptuk, hogy a kifejezés értéke divergál, míg adott $\alpha$ értékek mellett $1\,/\,2 \alpha$. Mivel $\alpha$ valós, így argumentuma $\pi\,/\, 2$ egészszámú többszöröse. Mivel a megoldás szerint ez kisebb kell legyen abszolútértékben, mint $\pi\,/\,2$, így értéke $0$, melyből következik egyben, hogy $\alpha$ pozitív. Tehát az integrál pozitív $\alpha$ értékek esetében lesz véges. Az $e^{- 2 \alpha r}$ függvényt ábrázolhatjuk egy grafikonon is így különböző $\alpha$ értékek esetén.

```python
def f(r, a):
    
    return np.exp(-2 * a * r)
```

```python
r_values = np.arange(0, 10, 0.1)
a_values = [0.1, 0.2, 0.5, 1, 1.5, 3]
```

```python
nrows = 1
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*10, nrows*8))

axislabelsize = 25
axisticksize = 15
axislegendsize = 18

colors = [
    'tab:red',
    'tab:green',
    'tab:orange',
    'tab:blue',
    'tab:purple',
    'tab:cyan'
]

for idx, a in enumerate(a_values):
    axes.plot(r_values, f(r_values, a), label='$\\alpha = {}$'.format(a),
              c=colors[idx], lw=3)

axes.set_xlabel('$r$', fontsize=axislabelsize)
axes.set_ylabel('$e^{-2 \\alpha r}$', fontsize=axislabelsize)
axes.tick_params(axis='both', which='major', labelsize=axisticksize)

axes.legend(loc='upper right', fontsize=axislegendsize)

plt.show()
```

A számlálóban szereplő kifejezést hasonló módon számíthatjuk, mint a fentieket:

$$
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{H} \phi
=
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{T} \phi
+
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{V} \phi
$$

A kinetikus energiát tartalmazó tag értéke - figyelembe véve, hogy $r$ miatt polárkoordinátákban kell a Laplace-operátorral történő operátorműveletet elvégeznünk, valamint, hogy $r \geq 0$:

$$
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{T} \phi
=
\int_{0}^{\infty} \text{d}r\, \left( e^{- \alpha r} \right)^{\ast} \cdot \left(\frac{\hbar^{2}}{2m} \Delta e^{- \alpha r} \right)
$$

A gömbi Laplace operátornak kizárólag az $r$ szerinti deriváltat tartalamzó tagja nem zérus, melyet hattatva a próbafüggvényre, az alábbi kifejezés adódik:

$$
\Delta e^{- \alpha r}
=
\frac{1}{r^{2}} \frac{\partial}{\partial r} \left( r^{2} \frac{\partial}{\partial r} e^{- \alpha r} \right)
=
\frac{1}{r^{2}} \frac{\partial}{\partial r} \left( - \alpha r^{2} \cdot e^{- \alpha r} \right)
=
\frac{1}{r^{2}} \left( - 2 \alpha r \cdot e^{- \alpha r} + \alpha^{2} r^{2} \cdot e^{- \alpha r} \right)
=
\frac{\alpha e^{- \alpha r} \left( \alpha r - 2\right)}{r}
$$

$$
\int_{0}^{\infty} \text{d}r\, \left( e^{- \alpha r} \right) \cdot \left(\frac{\hbar^{2}}{2m} \frac{\alpha e^{- \alpha r} \left( \alpha r - 2\right)}{r} \right)
=
\frac{\hbar^{2} \alpha}{2m}
\int_{0}^{\infty} \text{d}r\, \frac{e^{- 2 \alpha r} \left( \alpha r - 2\right)}{r}
$$

Érdekes lehet az integrandus értékét megint kiplottolni különböző $\alpha$ értékek esetén. 

```python
def f2(r, a):
    
    return np.exp(-2 * a * r) * (a * r - 2) / r
```

```python
r_values_s = np.arange(-0.002, 0.002, 0.0001)
r_values_l = np.arange(-4, 4, 0.0001)
a_values = [0.1, 0.2, 0.5, 1, 1.5, 3]
```

```python
nrows = 1
ncols = 2
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*10, nrows*8))

axislabelsize = 24
axisticksize = 15
axislegendsize = 18

colors = [
    'tab:red',
    'tab:green',
    'tab:orange',
    'tab:blue',
    'tab:purple',
    'tab:cyan'
]

for idx, a in enumerate(a_values):
    axes[0].plot(r_values_s, f2(r_values_s, a), label='$\\alpha = {}$'.format(a),
                 c=colors[idx], lw=3)
    
for idx, a in enumerate(a_values):
    axes[1].plot(r_values_l, f2(r_values_l, a), label='$\\alpha = {}$'.format(a),
                 c=colors[idx], lw=3)

for i in range(ncols):
    axes[i].set_xlabel('$r$', fontsize=axislabelsize)
    axes[i].set_ylabel('$\\frac{\\alpha e^{-2 \\alpha r} \\left( \\alpha r - 1 \\right)}{r}$', fontsize=axislabelsize)
    axes[i].tick_params(axis='both', which='major', labelsize=axisticksize)

    axes[i].legend(loc='upper right', fontsize=axislegendsize)

plt.show()
```

A függvény érthetően $r=0$-ban az $1\,/\,r$ függés miatt divergál, negatív $r$ értékekre pedig szintén nagyon hamar elszáll. A numerikus hibák miatt az $r=0$ közelében mutatott érték természetesen nem valós, csak egy afféle útmutató. Az integrált megint pl. a `sympy` segítségével értékelhetjük ki.

```python
r = sp.Symbol('r')
alpha = sp.Symbol('\\alpha')
```

```python
result = sp.integrate((sp.exp(-2 * alpha * r) * (alpha * r - 2)) / r, (r, 0, sp.oo))
result
```

Mely eredmény a `sympy` ismeretében arra utal, hogy az integrál értéke nem véges. Ez feltételezhetően nem helyes, ellenben a teljes fenti kifejezés kísértetiesen hasonlít a $\phi = e^{- \alpha x^{2}}$ próbafüggvénnyel végzett eredményhez. A fizikus megérzést alkalmazva tegyük fel, hogy a fenti integrál értéke pontosan $1$, és csupán a `sympy` nem képes megbírkózni vele. Így a végső eredmény az alábbi lesz:

$$
\int_{0}^{\infty} \text{d}r\, \phi^{\ast} \hat{T} \phi
=
\frac{\hbar^{2} \alpha}{2m}
$$

A potenciális energiát tartalmazó kifejezés az alábbi módon számítható:

$$
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{V} \phi
=
\int_{-\infty}^{\infty} \text{d}r\, \left( e^{- \alpha r} \right)^{\ast} \cdot \left(\frac{1}{2} k r^{2} e^{- \alpha r} \right)
\equiv
\int_{0}^{\infty} \text{d}r\, \left( e^{- \alpha r} \right)^{\ast} \cdot \left(\frac{1}{2} k r^{2} e^{- \alpha r} \right)
=
\frac{1}{2} k
\int_{0}^{\infty} \text{d}r\, r^{2} e^{- 2 \alpha r}
$$

```python
r = sp.Symbol('r')
alpha = sp.Symbol('\\alpha')
```

```python
result = sp.integrate(r**2 * sp.exp(-2 * alpha * r), (r, 0, sp.oo))
result
```

Hasonlóan a normalizációs faktornál kapottakhoz, az $\alpha$ lehetséges értékei alapján az ingerandus értéke $1\,/\,4 \alpha^{3}$. Így a potenciált tartalmazó tag értéke az alábbi:

$$
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{V} \phi
=
\frac{k}{8 \alpha^{3}}
$$

A fentieket aggregálva az $E_{\phi}$ értékére a következőket kapjuk:

$$
E_{\phi}
=
\frac{
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{T} \phi
+
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \hat{V} \phi
}
{
\int_{-\infty}^{\infty} \text{d}r\, \phi^{\ast} \phi
}
=
\frac{
\frac{\hbar^{2} \alpha}{2m} + \frac{k}{8 \alpha^{3}}
}
{
\frac{1}{2 \alpha}
}
=
\frac{\hbar^{2} \alpha^{2}}{m} + \frac{k}{4 \alpha^{2}}
$$

Ezt a függvényt is ábrázolhatjuk, most $\alpha$ értéke szerint, különböző $k$ paraméterekre. Itt vegyük $\hbar$ és $m$ értékét $1$-nek.

```python
def E_phi(a, k):
    
    return a**2 + k/(4 * a**2)
```

```python
a_values = np.arange(0.01, 10, 0.01)
k_values = [0.1, 0.2, 0.5, 1, 1.5, 3]
```

```python
nrows = 1
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*10, nrows*8))

axislabelsize = 25
axisticksize = 15
axislegendsize = 18

colors = [
    'tab:red',
    'tab:green',
    'tab:orange',
    'tab:blue',
    'tab:purple',
    'tab:cyan'
]

for idx, k in enumerate(k_values):
    axes.plot(a_values, E_phi(a_values, k), label='$k = {}$'.format(k),
              c=colors[idx], lw=3)

axes.set_ylim(-10,110)
    
axes.set_xlabel('$\\alpha$', fontsize=axislabelsize)
axes.set_ylabel('$E_{\\varphi}$', fontsize=axislabelsize)
axes.tick_params(axis='both', which='major', labelsize=axisticksize)

axes.legend(loc='lower right', fontsize=axislegendsize)

plt.show()
```

Látható, hogy a kapott $E_{\phi}$ függvénynek feltehetően csak egy minimuma van, az ábra paraméterezésében valahol $\alpha$ értékében $0$ és $1$ között. Ennek minimumát az $E_{\phi}$, $\alpha$ szerinti deriválásával kaphatjuk, megkeresve annak zérus értékének helyét. Deriváljunk egyszer `sympy`-al is, ha már Jupyter Notebookban dolgozom.

```python
alpha = sp.Symbol('\\alpha')
result = sp.diff(alpha**2 + 1/alpha**2, alpha)
result
```

Tehát $E_{\phi}$, $\alpha$ szerinti deriváltja az alábbi:

$$
\frac{d E_{\phi}}{d \alpha}
=
\frac{2 \hbar^{2} \alpha}{m} - \frac{k}{2 \alpha^{3}}
$$

Ezt is ábrázolhatjuk egy grafikonon:

```python
def E_phi_diff(a, k):
    
    return 2 * a - k/(2 * a**3)
```

```python
nrows = 1
ncols = 1
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(ncols*10, nrows*8))

axislabelsize = 25
axisticksize = 15
axislegendsize = 18

colors = [
    'tab:red',
    'tab:green',
    'tab:orange',
    'tab:blue',
    'tab:purple',
    'tab:cyan'
]

for idx, k in enumerate(k_values):
    axes.plot(a_values, E_phi_diff(a_values, k), label='$k = {}$'.format(k),
              c=colors[idx], lw=3)

axes.set_ylim(-20,40)
    
axes.set_xlabel('$\\alpha$', fontsize=axislabelsize)
axes.set_ylabel('$\\frac{d E_{\\varphi}}{d \\alpha}$', fontsize=axislabelsize)
axes.tick_params(axis='both', which='major', labelsize=axisticksize)

axes.legend(loc='lower right', fontsize=axislegendsize)

plt.show()
```

Keressük azt az $\alpha$ értéket, ahol a függvénynek zérushelye van. A számítás a következő. Keressük az alábbit:

$$
\frac{2 \hbar^{2} \alpha}{m} - \frac{k}{2 \alpha^{3}}
=
0
$$

Áterndezve $\alpha$-ra az egyenletet:

$$
\alpha_{\text{min}}^{4} = \frac{km}{4 \hbar^{2}}
$$

$$
\alpha_{\text{min}}^{2} = \frac{\sqrt{km}}{2 \hbar}
$$

A kapott egyenlet jobb oldala megegyezik az útmutatónak kapott leírásokban/videókban, $\alpha$-ra kapott eredménnyel, a bal oldalon $\alpha_{\text{min}}$ helyett viszont $\alpha^{2}_{\text{min}}$ szerepel. Feltételezem, hogy a "Mit tapasztalunk, miért?" kérdésre ezt a jelenséget kéne megválaszolni. Ebben nem vagyok semennyire sem biztos, azonban jelentős különbséget csupán az jelentette a számításban, hogy itt polárkoordinátákban dolgoztunk, az $r$ értéke $0$ és $\infty$ között mozgott, valamint a próbafüggvényben csupán elsőfokúnak vettük a leírásokban másodfokúként szereplő tagot.
