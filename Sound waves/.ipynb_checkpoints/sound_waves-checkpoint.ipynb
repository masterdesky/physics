```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation
```

```python
h = 1
w = 5
n = 1000
X = np.random.rand(n) * w
Y = np.random.rand(n) * h
```

### Generate points

```python
fig, axes = plt.subplots(figsize=(w*10, h*10))

axes.scatter(X, Y)

plt.show()
```

### Generate wave amplitudes

```python
fig, axes = plt.subplots(figsize=(w*10, h*10))

frames = 200
wave_A = 0.1

line, = axes.plot([], [], lw=3)
axes.set_xlim(0, w)
axes.set_ylim(- wave_A - wave_A/10, wave_A + wave_A/10)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = np.linspace(0, w, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i)) * wave_A
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                     frames=frames, interval=20, blit=True)

#anim.save('sine_wave.gif', writer='ffmpeg')

Writer = animation.writers['ffmpeg']
writer = Writer(fps=20, bitrate=28000,
                metadata=dict(artist='Balazs Pal'),)
anim.save('sine_wave.mp4', dpi=50, writer=writer)
```

### Plot waves

```python
wave_A = 0.1
x = np.linspace(0, w, 1000)
y = np.sin(2 * np.pi * x) * wave_A
```

```python
fig, axes = plt.subplots(figsize=(w*10, h*10))

axes.scatter(X+y, Y)

plt.show()
```

```python
fig, axes = plt.subplots(figsize=(w*10, h*10))

frames = 200
wave_A = 0.1

scat = axes.scatter([],[])
axes.set_xlim(0, w)
axes.set_ylim(0, h)

def animate(i):
    x = np.linspace(0, w, n)
    y = np.sin(2 * np.pi * (x - 0.01 * i)) * wave_A

    scat.set_offsets(np.dstack((X+y,Y))[0])
    return scat,

anim = FuncAnimation(fig, animate,
                     frames=frames, interval=20, blit=True)

#anim.save('sine_wave.gif', writer='ffmpeg')

Writer = animation.writers['ffmpeg']
writer = Writer(fps=20, bitrate=28000,
                metadata=dict(artist='Balazs Pal'),)
anim.save('perturbation.mp4', dpi=50, writer=writer)
```

```python
t = np.linspace(0, w, n)
x = np.sin(2 * np.pi * t) * wave_A
v = np.cos(2 * np.pi * t) * wave_A
```

```python
plt.figure(figsize=(w*5,h*5))
plt.plot(t,x)
plt.plot(t,v)
```
